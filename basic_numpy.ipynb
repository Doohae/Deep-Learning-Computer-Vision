```python
# 결과 출력 양식 별도 지정 함수
def print_result(SIZE, time_add_lst, time_dot_lst, time_add_np, time_dot_np, addition_lst, product_lst, addition_np, product_np):
  HALF = int(SIZE/2)
  FULL = int(SIZE-1)
  print('-------------------------------------------------------')
  print('SIZE = ' + str(SIZE) + ' HALF = ' + str(HALF) + ' FULL = ' + str(FULL))
  print('-------------------------------------------------------')
  print('Addition values at 0, HALF, FULL = ' + str(addition_lst[0]) +', ' + str(addition_lst[HALF]) + ', ' + str(addition_lst[FULL]))
  print('List addition time [ms] = ' + str(time_add_lst))
  print('-------------------------------------------------------')
  print('List dot product = ' + str(product_lst))
  print('List dot product time [ms] = ' + str(time_dot_lst))
  print('-------------------------------------------------------')
  print('Addition values at 0, HALF, FULL = ' + str(addition_np[0]) + ', ' + str(addition_np[HALF]) + ', ' + str(addition_np[FULL]))
  print('NumPy addition time [ms] = ' + str(time_add_np))
  print('-------------------------------------------------------')
  print('NumPy dot product = ' + str(product_np))
  print('NumPy dot product time [ms] = ' + str(time_dot_np))
  print('-------------------------------------------------------')
  print('\n\n')
```

```python
import numpy as np
import time

SIZE_list = [100, 10000, 1000000]

######### Step 1 : SIZE_list에 측정 타겟 사이즈 100, 10000, 1000000을 넣고 loop 돌려 실행 ###########

for SIZE in SIZE_list:

  L1 = range(SIZE)
  L2 = range(SIZE)
  A1 = np.arange(SIZE)
  A2 = np.arange(SIZE)
  # -------------------------------------------------------
######### Step 2-1 : SIZE에 대한 list addition(loop) 실행 ###########
  start_add_lst = time.time()

  addition_lst = []
  for i in range(len(L1)):
    addition_lst.append(L1[i]+L2[i])

  end_add_lst = time.time()

  time_add_lst = (end_add_lst - start_add_lst)*1000

  # -------------------------------------------------------
######### Step 2-2 : SIZE에 대한 list dot product(loop) 실행 ###########
  start_dot_lst = time.time()

  product_lst = 0
  for i in range(len(L1)):
    product_lst += L1[i]*L2[i]

  end_dot_lst = time.time()

  time_dot_lst = (end_dot_lst - start_dot_lst)*1000

  # -------------------------------------------------------
######### Step 2-3 : SIZE에 대한 numpy array addition 실행 ###########
  start_add_np = time.time()

  addition_np = A1 + A2

  end_add_np = time.time()

  time_add_np = (end_add_np - start_add_np)*1000

  # -------------------------------------------------------
######### Step 2-4 : SIZE에 대한 numpy array dot product 실행 ###########
  start_dot_np = time.time()

  product_np = np.dot(A1, A2)

  end_dot_np = time.time()

  time_dot_np = (end_dot_np - start_dot_np)*1000

  # -------------------------------------------------------
######### Step 3 : Step2를 통해 얻은 측정 시간들을 함수 호출을 통해 양식에 맞게 출력 ###########
  print_result(SIZE, time_add_lst, time_dot_lst, time_add_np, time_dot_np, addition_lst, product_lst, addition_np, product_np)
```
